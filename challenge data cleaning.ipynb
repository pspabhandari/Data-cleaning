```python
#non-standard for NA
missing_values = ["n/a", "na", "NaN", "--", 'inf']
wellcome = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', na_values = missing_values)
wellcome_c = wellcome.copy()
wellcome_c.head()
```

```python
#changing COST column name to make it easier
wellcome_c.rename(columns = {'COST (�) charged to Wellcome (inc VAT when charged)': 'Cost Charged to Wellcome in £'}, inplace = 'True' )
```

```python
wellcome_c.head()
```

```python
#splitting character � and integers to delete �
welcome_split = wellcome_c['Cost Charged to Wellcome in £'].str.split('�', expand = True)
```

```python
#splitting and getting integers in 1
wellcome_c['Cost Charged to Wellcome in £']=welcome_split.get(1)
```

```python

wellcome_c.head()
```

```python
wellcome_c['Cost Charged to Wellcome in £'] = pd.to_numeric(wellcome_c['Cost Charged to Wellcome in £'])
print(wellcome_c['Cost Charged to Wellcome in £'].dtypes)
```

```python
wellcome_c['Cost($)'] = wellcome_c['Cost Charged to Wellcome in £'] * 1.24
wellcome_c['Cost($)'] = round(wellcome_c['Cost($)'], 2)
```

```python
wellcome_c.dropna(inplace=True)
wellcome_c.head()
```

```python
#counting total journal duplicates
wellcome_c['Journal title'].value_counts().head()
```

```python
#replacing journal title to proper one for same journal writing in different ways
wellcome_c['Journal title'].replace(to_replace =['PLoS ONE', 'PLOS ONE','PLOS One', 'PLoS 1', 'Plos ONE', 'PloS ONE',
                                              'PLoSONE', 'PLosONE'], value = 'PLoS One',inplace = True ) 
wellcome_c['Journal title'].replace(to_replace =['Journal Biological Chemistry', 'J Biol Chem.','JOURNAL OF BIOLOGICAL CHEMISTRY', 'The Journal of Biological Chemistry', 'Jnl Biological Chemistry',
                                              'Journal of Biol Chem', 'Biologicial Chemistry', 'J Biol Chemistry', 'Journal of biological chemistry', 'Biochem Journal', 'The journal of Biological Chemistry'], 
                                  value = 'Journal of Biological Chemistry',inplace = True ) 
wellcome_c['Journal title'].replace(to_replace =['Nucleic Acid Research', 'NUCLEIC ACIDS RESEARCH',
                                               'Nucleic Acids Research\n'], value = 'Nucleic Acids Research',inplace = True )
wellcome_c['Journal title'].replace(to_replace =['PNAS', 'PNAS (Proceedings of the National Academy of Sciences of the United States of America',
                                               'PNAS USA', 'Proceedings of the National Academy of Sciences of the USA',
                                              'Proc Natl Acad Sci U S A.', 'Proceedings of the national Academy of Sciences',
                                              'Proceedings of the National Academy of Sciences (PNAS)',
                                              'Proceedings of the National Academy of Sciences of the United States of America',
                                              'Proceddings of the National Academy of Sciences of USA'],
                                  value = 'Proceedings of the National Academy of Sciences',inplace = True )
wellcome_c['Journal title'].replace(to_replace =['PLoS Negected Tropical Diseases', 'PLOS Neglected Tropical Diseases'], 
                                  value = 'PLoS Neglected Tropical Diseases',inplace = True )
wellcome_c['Journal title'].replace(to_replace =['Human molecular genetics', 'Human Molecular Genetics Online',
                                              'Human Mol Genetics Online'], 
                                  value = 'Human Molecular Genetics',inplace = True )
wellcome_c['Journal title'].replace(to_replace =['Nature communications', 'NATURE COMMUNICATIONS'], 
                                  value = 'Nature Communications',inplace = True )
wellcome_c['Journal title'].replace(to_replace =['PLOS Pathogens', 'Plos Pathogens'
                                              ], value = 'PLoS Pathogens',inplace = True ) 
wellcome_c['Journal title'].replace(to_replace =['PLOS Genetics', 'Plos Genetics','PLoS GENETICS'
                                              ], value = 'PLoS Genetics',inplace = True ) 
wellcome_c['Journal title'].replace(to_replace =['Neuroimage', 'Neurolmage','Neuroimage: Clinical', 'Neuroimage: clinical',
                                               'NeuroImage: Clinical'
                                              ], value = 'NeuroImage',inplace = True ) 

#five common Journal article with its total count
wellcome_c['Journal title'].value_counts().head()

```
